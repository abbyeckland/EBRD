# Figure 4
## Repeat sediment-water and bed material sample data

```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from matplotlib.dates import DateFormatter, MonthLocator
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
import matplotlib as mpl
```

### First import the data
#### 1. Narrows 15-minute interval water discharge, sampling period (July 2021-October 2022)

```python
# Load data from file
na = pd.read_csv('../Hydro_Data/USGS/Narrows_08359500/08359500_07_01_21-10_03_22.csv', skiprows=28)

# Rename columns
na = na.rename(columns={'101298_00060_cd':'Qw (ft^3/s)'})

# convert datetime column to to_datetime pandas object
na['DateTime'] = pd.to_datetime(na['datetime'])

# Linearly interpolate between cells that contain missing water discharge data
na['Qw (ft^3/s)'] = na['Qw (ft^3/s)'].interpolate()

# Convert water discharge units from imperial (f^3/s) to metric (m^3/s)
na['Qw (m^3/s)'] = na['Qw (ft^3/s)']*0.028316846592

# Drop original Qw column
na = na.drop(['Qw (ft^3/s)', 'datetime'], axis=1)

# Fill any null values with 0
na = na.fillna(0)

# Preview data
na
```

#### 2. SSC and OM% Field Data

```python
# Load data from file
ssc = pd.read_csv('../Field_Data/Fig4_SSC-OM.csv')

# Convert datetime column to to_datetime pandas object
ssc['DateTime']= pd.to_datetime(ssc['DateTime'])

# Conver SSC in mg/l to kg/m^3
ssc['SSC (kg/m3)'] = (ssc['SSC (mg/l)']*1e-6)/1e-3

# Preview data
ssc
```

#### 3. Grain size distributions for suspended load and bed material
**First the grain size distributions for the suspended sediment**

```python
# Import data from file
gs = pd.read_csv('../Field_Data/Fig4_SS-GSD.csv')

# Convert datetime column to to_datetime pandas object
gs['DateTime']= pd.to_datetime(gs['DateTime'])

# Preview data
gs.head()
```

Calculate the mean D10, D50, and D90 among all samples from each season

```python
# July 2021 samples
jul_d10 = np.mean(gs['D10'][0:8])
jul_d50 = np.mean(gs['D50'][0:8])
jul_d90 = np.mean(gs['D90'][0:8])

# March 2022 samples
mar_d10 = np.mean(gs['D10'][9:14])
mar_d50 = np.mean(gs['D50'][9:14])
mar_d90 = np.mean(gs['D90'][9:14])

# May 2022 samples
may_d10 = np.mean(gs['D10'][14:18])
may_d50 = np.mean(gs['D50'][14:18])
may_d90 = np.mean(gs['D90'][14:18])

# September 2022 samples
sep_d10 = np.mean(gs['D10'][18:22])
sep_d50 = np.mean(gs['D50'][18:22])
sep_d90 = np.mean(gs['D90'][18:22])

# Combine data into arrays to get D10, D50, and D90 values for each sampling campaign
d10 = np.array([jul_d10, gs['D10'][8], mar_d10, may_d10, sep_d10])
d50 = np.array([jul_d50, gs['D50'][8], mar_d50, may_d50, sep_d50])
d90 = np.array([jul_d90, gs['D90'][8], mar_d90, may_d90, sep_d90])
time = np.array([gs['DateTime'][0], gs['DateTime'][8], gs['DateTime'][9], gs['DateTime'][14], gs['DateTime'][18]])
```

**Second the grain size distributions for the bed material data**

```python
# Import the data
bed = pd.read_csv('../Field_Data/Fig4_Bed-GSD.csv')

# Convert datetime column to datetime pandas object
bed['DateTime']= pd.to_datetime(bed['DateTime'])

# Preview data
bed.head()
```

Calculate the mean D10, D50, and D90 among all samples from each season

```python
# July 2021 samples
jul_d10_bed = np.mean(bed['D10'][0:6])
jul_d50_bed = np.mean(bed['D50'][0:6])
jul_d90_bed = np.mean(bed['D90'][0:6])

# March 2022 samples
mar_d10_bed = np.mean(bed['D10'][7:13])
mar_d50_bed = np.mean(bed['D50'][7:13])
mar_d90_bed = np.mean(bed['D90'][7:13])

# September 2022 samples
sep_d10_bed = np.mean(bed['D10'][13:16])
sep_d50_bed = np.mean(bed['D50'][13:16])
sep_d90_bed = np.mean(bed['D90'][13:16])

# Combine data into arrays to get D10, D50, and D90 values for each sampling campaign
d10_bed = np.array([jul_d10_bed, bed['D10'][6], mar_d10_bed, sep_d10_bed])
d50_bed = np.array([jul_d50_bed, bed['D50'][6], mar_d50_bed, sep_d50_bed])
d90_bed = np.array([jul_d90_bed, bed['D90'][6], mar_d90_bed, sep_d90_bed])
time_bed = np.array([bed['DateTime'][0], bed['DateTime'][6], bed['DateTime'][7], bed['DateTime'][13]])
```

## Make Figure 4A

```python
# Build the figure
fig = plt.figure(figsize = (12.95, 5))
ax = fig.add_subplot()

# Add colorful shading for different seasons
plt.axvspan(pd.Timestamp('2021-07-01'), pd.Timestamp('2021-10-01'), color='peru', alpha = 0.2)
plt.axvspan(pd.Timestamp('2021-10-01'), pd.Timestamp('2022-04-01'), color='royalblue', alpha = 0.2)
plt.axvspan(pd.Timestamp('2022-04-01'), pd.Timestamp('2022-07-01'), color='green', alpha = 0.2)
plt.axvspan(pd.Timestamp('2022-07-01'), pd.Timestamp('2022-10-01'), color='peru', alpha = 0.2)

# Add water discharge data on left axils
lns1 = ax.plot(na['DateTime'], na['Qw (m^3/s)'], color='grey', linewidth=1, label = 'Discharge')
ax.xaxis.set_tick_params(labelsize=19, length=5, width=1)

# Make second y axis and add SSC sample data
ax2 = ax.twinx()
lns2 = ax2.plot(ssc['DateTime'], ssc['SSC (kg/m3)'], linestyle='None', color = 'tab:brown', marker = 'o', markerfacecolor='None', 
                markeredgewidth=2.2, markersize=18,markeredgecolor='saddlebrown', label = 'SSC')

# Add hyperpyncal threshold line
ax2.axhline(y=1e0, color='darkred', linewidth = 1.5, linestyle='--', label='Hyperpycnal threshold')

# Build legend
ax.legend(loc='lower left', fontsize=18, facecolor='white', borderaxespad=0.1)

# Format plot
date_form = DateFormatter("%m/%y")
ax.xaxis.set_major_formatter(date_form)
ax.set_ylabel('Water Discharge (m$^3$/s)', size = 23)
ax.set_yscale('log')
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f')), ax.yaxis.set_tick_params(labelsize=19, length=5, width=1)
ax.set_xlim(pd.Timestamp('2021-07-01'), pd.Timestamp('2022-10-01')), ax.set_ylim(0.5e0,100)


ax2.legend( loc='lower center', fontsize=18, facecolor='white', borderaxespad=0.1)
ax2.set_ylabel('SSC (kg/m$^3$)', size = 23, color='k'), ax2.yaxis.set_label_coords(1.05, .5)
ax2.margins(0.01), ax2.set_yscale('log')
ax2.yaxis.set_major_formatter(FormatStrFormatter('%.0f')), ax2.yaxis.set_tick_params(labelsize=19, length=5, width=1)
ax2.set_ylim(1e-1,10e1)

plt.tight_layout()

# Save figure to folder
plt.savefig('Saved_Figures/Figure4A', dpi=150)

plt.show()
```

## Make Figure 4E

```python
# Build the figure
fig = plt.figure(figsize = (16.2, 4.5))
ax = fig.add_subplot()

# Add colorful shading for different seasons
plt.axvspan(pd.Timestamp('2021-07-01'), pd.Timestamp('2021-10-01'), color='peru', alpha = 0.2)
plt.axvspan(pd.Timestamp('2021-10-01'), pd.Timestamp('2022-04-01'), color='royalblue', alpha = 0.2)
plt.axvspan(pd.Timestamp('2022-04-01'), pd.Timestamp('2022-07-01'), color='green', alpha = 0.2)
plt.axvspan(pd.Timestamp('2022-07-01'), pd.Timestamp('2022-10-01'), color='peru', alpha = 0.2)

# Add water discharge data to left axis
ax.plot(na['DateTime'], na['Qw (m^3/s)'], color='k', alpha=0.5, linewidth=1, label = 'Discharge')
ax.xaxis.set_tick_params(labelsize=16, length=5, width=1)

# Add second y axis
ax2 = ax.twinx()

# Add size class boundaries for sand, silt, and clay
ax2.hlines(y=2e3, xmin = na['DateTime'][0], xmax= na['DateTime'][43877], colors='peru', linestyle= ':', label='Sand', linewidth=2.5)
ax2.hlines(y=63, xmin = na['DateTime'][0], xmax= na['DateTime'][43877], colors='brown', linestyle= '-.', label='Silt', linewidth=1.5)
ax2.hlines(y=3.9, xmin = na['DateTime'][0], xmax= na['DateTime'][43877], colors='chocolate', linestyle= '--', label='Clay', linewidth=1.5)

# Add vertical line that connects D10 and D90 for the SS data 
ax2.vlines(x=time[0], ymin=d10[0], ymax=d90[0], colors='saddlebrown')
ax2.vlines(x=time[1], ymin=d10[1], ymax=d90[1], colors='saddlebrown')
ax2.vlines(x=time[2], ymin=d10[2], ymax=d90[2], colors='saddlebrown')
ax2.vlines(x=time[3], ymin=d10[3], ymax=d90[3], colors='saddlebrown')
ax2.vlines(x=time[4], ymin=d10[4], ymax=d90[4], colors='saddlebrown')

# Add all D10, D50, and D90 data for the SS samples
ax2.plot(time, d90, linestyle='None', color='saddlebrown', marker = '_', markersize=18, markeredgewidth=2, label = 'SS: D90')
ax2.plot(time, d50, linestyle='None', color='saddlebrown', marker = '_', markersize=16, markeredgewidth=4, label = 'SS: D50')
ax2.plot(time, d10, linestyle='None', color='saddlebrown', marker = '_',  markersize=8, markeredgewidth=2, label = 'SS: D10')

# Add verttical line that connects d10 and D90 for the bed material data
ax2.vlines(x=time_bed[0], ymin=d10_bed[0], ymax=d90_bed[0], colors='peru')
ax2.vlines(x=time_bed[1], ymin=d10_bed[1], ymax=d90_bed[1], colors='peru')
ax2.vlines(x=time_bed[2], ymin=d10_bed[2], ymax=d90_bed[2], colors='peru')
ax2.vlines(x=time_bed[3], ymin=d10_bed[3], ymax=d90_bed[3], colors='peru')

# Add all D10, D50, and D90 data for the bed samples
ax2.plot(time_bed, d90_bed, linestyle='None', color='peru', marker = '_', markersize=18, markeredgewidth=2, label = 'BM: D90')
ax2.plot(time_bed, d50_bed, linestyle='None', color='peru', marker = '_', markersize=16, markeredgewidth=4, label = 'BM: D50')
ax2.plot(time_bed, d10_bed, linestyle='None', color='peru', marker = '_',  markersize=8, markeredgewidth=2, label = 'BM: D10')

# Format plot
ax.legend(fontsize=14, facecolor='white', loc='lower left', borderaxespad=0.1)
date_form = DateFormatter("%m/%y")
ax.xaxis.set_major_formatter(date_form)
ax.set_ylabel('Water Discharge (m$^3$/s)', size = 20, color='k')
ax.set_yscale('log')
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax.yaxis.set_tick_params(labelsize=16, length=5, width=1)
ax.set_xlim(pd.Timestamp('2021-07-01'), pd.Timestamp('2022-10-01'))
ax.set_ylim(0.5e0,100)
ax.yaxis.set_label_coords(-0.04,0.5)

ax2.legend(fontsize=14, facecolor='white', loc='lower center', ncol=3, borderaxespad=0.1)
ax2.set_ylabel('Grain size ($\mu $m)', size = 20, color='k')
ax2.set_ylim(1e-1,3e3)
ax2.margins(0.01)
ax2.set_yscale('log')
ax2.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax2.yaxis.set_tick_params(labelsize=16, length=5, width=1)
ax2.yaxis.set_label_coords(1.04,0.5)

plt.tight_layout()

# Save figure to folder
plt.savefig('Saved_Figures/Figure4E', dpi=150)
plt.show()
```
